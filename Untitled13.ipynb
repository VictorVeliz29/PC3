Parte II. Recuperación de Datos

```python
import requests

url = "https://rickandmortyapi.com/api/character"

data = requests.get(url).json()

personajes = data["results"]

print("Cantidad obtenida:", len(personajes))
```

```python
for p in personajes[:15]:
    print(p["name"], p["status"], p["species"])
```

```python
import re

for p in personajes[:15]:

    url_personaje = p["url"]

    patron = r"(\d+)$"

    resultado = re.search(patron, url_personaje)

    if resultado:
        print(p["name"], "-> ID:", resultado.group(1))
```

Parte III. DataFrame

```python
import pandas as pd

datos = []

for p in personajes:

    datos.append({
        "Nombre": p["name"],
        "Estado": p["status"],
        "Especie": p["species"],
        "Genero": p["gender"]
    })

df = pd.DataFrame(datos)

print(df.head())
```

```python
df["Estado"] = df["Estado"].str.upper()
df["Especie"] = df["Especie"].str.strip()
```

```python
print(df["Estado"].value_counts())
print(df["Especie"].value_counts().head(10))
print(df["Genero"].value_counts())
```

Parte IV. Visualización

```python
import matplotlib.pyplot as plt

df["Especie"].value_counts().head(10).plot(kind="bar")

plt.title("Top 10 especies")
plt.xlabel("Especie")
plt.ylabel("Cantidad")

plt.show()
```
